```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
# tex include
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
# matplotlib.rcParams["text.latex.preamble"].append(r'\usepackage[dvips]{graphicx}\usepackage{xfrac}')
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$\sin(x)$')
plt.plot(x, y2, label=r'$\cos(x)$')

# Add title and labels rendered with LaTeX
plt.title(r'$\text{Trigonometric Functions: } \sin(x) \text{ and } \cos(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
